- Apa itu machine learning?
- Apa itu data feature dan data target?
- Apa Perbedaan Supervised Learning dan Unsupervised Learning?
- Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!
- Apa perbedaan Hyperparameter dan Parameter?
- Sebutkan Langkah-langkah dalam mengaplikasikan algoritma apapun dalam machine learning!

- Machine learning adalah salah satu bidang keilmuan yang memberikan komputer kemampuan untuk belajar, tanpa harus diprogram secara eksplisit


- Data feature adalah variabel yang memprediksi suatu kejadian atau target. Sedangkan data target adalah hasil dari prediksi tersebut


- Perbedaannya terdapat pada labelnya. Supervised learning memiliki kolom/data label serta outputnya jelas, sedangkan unsupervised learning tidak memiliki kolom/data label serta outputnya tidak jelas.


- Jenis-jenis Supervised Learning dan Perbedaannya

```
Jenis supervised learning
1. Klasifikasi
2. Regresi

Perbedaannya ada dioutput hasil prediksi. Output klasifikasi berupa class label/nilai diskrit, sedangkan output regresi adalah nilai kontinu.
```

- Hyperparameter adalah bagian dari parameter. Parameter belum tentu hyperparameter. Perbedaannya

```
1. Hyperparameter didefinisikan diawal pembuatan objek model, sedangkan parameter dihasilkan diakhir
2. Hyperparameter tidak memiliki hubungan dengan model, sedangkan parameter memiliki hubungan dengan model. Hyperparameter adalah suatu variabel yang didefinisikan untuk mengoptimasi model, parameter adalah sesuatu yang ada didalam model
```

- Langkah untuk mengaplikasikan machine learning

```
1. Pilih model atau algoritma yang akan digunakan. (contohnya linear regression, decision tree, logistic decision)
2. Pilih hyperparameter ketika membuat objek model
3. Pisahkan data feature dan data target. Pastikan input berupa matrix, bukan vektor (untuk sklearn).
4. Perintahkan model untuk mempelajari data dengan menggunakan method .fit()
5. Aplikasikan hasil belajar/model ke dalam test data menggunakan method .predict() (untuk SL) atau .transform() (untuk USL)
```

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

````python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

# input harus matriks
x_matriks = x[:, np.newaxis]
x_new_matriks = x_new[:, np.newaxis]
````

Contoh jawaban hasil pembuatan model dengan algoritma DecisionTreeRegressor (jawaban tidak harus sama):

Contoh jawaban parameters (tidak harus sama):

- n features : 1
  
Tree Graph :

````python
from sklearn.tree import DecisionTreeRegressor as dtr, plot_tree
#buat model dan tentukan hiperparameter
model = dtr(max_depth=4, min_samples_leaf=5, max_leaf_nodes=7, random_state=2)

# fit / proses belajar
model.fit(x_matriks, y)

# aplikasikan hasil belajar
# data awal
y_train = model.predict(x_matriks)
# data tes
y_test = model.predict(x_new_matriks)

# plot figure
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x,y, label='data awal')
ax.scatter(x, y_train, label='Hasil prediksi terhadap data training')
ax.plot(x_new, y_test, label='Hasil prediksi terhadap data baru')
ax.set_ylabel('Y-Value')
ax.set_xlabel('X-Value')
ax.set_title('Decision Tree Regressor')
fig.legend()
plt.show()

# plot_tree(model)
plottree, ax_tree = plt.subplots(figsize=(12,8))
ax_tree = plot_tree(model)
plt.show()
# parameter
print('Parameter dari decision tree regressor')
print('max features: ',model.max_features)
print('n features: ',model.n_features_)
print('n outputs: ',model.n_outputs_)
print('feature importances: ',model.feature_importances_)
````

````python
from sklearn.ensemble import RandomForestRegressor as rf
# buat objek model dan tentukan hiperparameter
model_rf = rf(max_depth=2, n_estimators=150)

# fitting / proses belajar
model_rf.fit(x_matriks, y)

# aplikasikan
y_train = model_rf.predict(x_matriks)
y_test = model_rf.predict(x_new_matriks)

# plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x,y, label='data awal')
ax.scatter(x, y_train, label='Hasil prediksi terhadap data training')
ax.plot(x_new, y_test, label='Hasil prediksi terhadap data baru')
ax.set_ylabel('Y-Value')
ax.set_xlabel('X-Value')
ax.set_title('Random Forest Regressor')
fig.legend()
````

````python
# Code here
from sklearn.linear_model import LinearRegression as linr
model_lr = linr(fit_intercept=True)


# model mempelajari data x_matriks supaya hasilnya serupa y
model_lr.fit(x_matriks,y)

# lihat output trainingnya
y_train = model_lr.predict(x_matriks)

# aplikasikan model terhadap data lain
y_test = model_lr.predict(x_new_matriks)

# plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x, y, color='g', label='Data Asli')
ax.plot(x, y_train, color='b', label='Hasil prediksi terhadap data training')
ax.plot(x_new, y_test, color='r', label='Hasil prediksi terhadap data baru')

ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
ax.set_title('Plot Linear Regression')
fig.legend()
plt.show()
# parameter
print('n feature :',model_lr.coef_[0])
print('intercept :',model_lr.intercept_)
````
