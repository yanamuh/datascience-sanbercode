# Soal 2: Facet Grid

Gunakan fungsi hexbin berikut untuk diterapkan pada mapping dari facetgridnya sehingga dihasilkan output sesuai expected.

```
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
```


````python
#code here
# sns.set_style('whitegrid')
data2 = df[['carat','cut','price']]
# fig2, ax = plt.subplots(figsize=(16,8), dpi=100)
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
with sns.axes_style('dark'):
    grid = sns.FacetGrid(data=data2, col='cut', hue='cut', height=4, sharex=True, sharey=True) # hue=cut.count
grid.map(hexbin, 'carat','price', extent=[0,3,0,20000]) #axis keseluruhan
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)
data3 = df[['carat','depth','table','price','cut']]
sns.pairplot(data3, hue='cut', markers=['+','o','s','D','x'], corner=True)
#code here
````
