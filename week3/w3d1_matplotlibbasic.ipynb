<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# figure dan axis kosong tanpa data

fig = plt.figure()
ax = fig.add_subplot()
# plt.ylabel(' ')
plt.show()
```

```python
# eksplisit
fig_1b = plt.figure()
ax_1b = fig_1b.add_subplot()
ax_1b.plot(x,y)
plt.show()
```

```python
# implisit
plt.plot(x,y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(x,y, x,y2, x,y3)
plt.show() # or fig.show()

# save figure
fig.savefig('sin.png')

# plt.plot(x,y)
# plt.plot(x,y2)
# plt.plot(x,y3)
```
