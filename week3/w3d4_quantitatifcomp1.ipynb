<h3> Preparing data </h3>

````python
# code here
# ubah tipe data tiap kolom
for col in list(df.columns):
    try:
        df[col] = df[col].astype(float) #try to set it as float
    except:
        df[col] = df[col].astype(str) #if its fail, try set it as str
df.dtypes
````

<b> <i> Beberapa kolom yang berisi data tipe numerik, tidak dapat dikonversi ke tipe float karena mengandung nilai '-' yang berupa string. Untuk itu, ubah dulu '-' menjadi NaN dengan menggunakan <b> <i> ``` np.NaN```

````python
# replace '-' menjadi np.NaN
df.replace('-', np.NaN, inplace=True)
df.isna().sum()
````

<i><b> Isi data NaN tersebut dengan menggunakan ```.fillna()``` dengan metode ```ffill```.
Metode `ffill` akan mengganti nilai NaN sama dengan nilai pada row sebelumnya di kolom yang sama

````python
df.fillna(method='ffill', inplace=True)
df.isna().sum()
````

````python
# ubah tipe data
for col in list(df.columns):
    try:
        df[col] = df[col].astype(float) #try to set it as float
    except:
        df[col] = df[col].astype(str) #if its fail, try set it as str
df.dtypes
````

<h3> Visualisasi data </h3>

```
Scatter plot dengan
x = HumidyAvgPercent
y = DewPointAvgF
color = TempAvgF
size = WindAvgMPH
```

````python
# Scatter plot data
fig, ax = plt.subplots(figsize=(16,8))
ax_ = ax.scatter(x=df.HumidityAvgPercent, y=df.DewPointAvgF, c=df.TempAvgF, s=20*df.WindAvgMPH, cmap='coolwarm')
ax.set_xlabel('Humidity Average (%)')
ax.set_ylabel('Dew Point Average (F)')
ax.set_title('Austin Weather')
cax = fig.colorbar(ax_) # fig.colorbar(mappable, ax=) mappable- data yg mau diambil, ax- colorbarnya mau ditaruh dimana
cax.ax.set_ylabel('Temperature Average (F)')
plt.show()
````

<h3> Insight </h3>

- Dew point memiliki korelasi linear yang kuat dengan temperature. Temperature yang semakin tinggi menyebabkan dew point meningkat.

- Dew point tidak memiliki korelasi linear dengan wind

- Humidity memiliki korelasi positif dengan dew point, secara umum humidity akan meningkat seiring dengan meningkatnya dew point. 

- Antara humidity dengan temperatur tidak memiliki korelasi. Meskipun temperatur meningkat, humiditinya tidak terlalu berubah

- Antara humidity dan wind pun tidak memiliki korelasi linear. Hal ini dapat terlihat dari ukuran data yang tidak membesar/mengecil seiring dengan peningkatan/penurunan humidity

- Semakin besar humidity, nilai dew point dan temperature cenderung sama.

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

````python
#code here
plt.style.use('ggplot')
datax = data.transpose()

# data
labels = datax.index
x = np.arange(0,8,2)
y = datax['Action']
y2 = datax['Adventure']
y3 = datax['Fighting']
y4 = datax['Misc']
y5 = datax['Platform']
y6 = datax['Puzzle']
y7 = datax['Racing']
y8 = datax['Role-Playing']
y9 = datax['Shooter']
y10 = datax['Simulation']
y11 = datax['Sports']
y12 = datax['Strategy']

# plot
fig3, ax3 = plt.subplots(figsize=(16,8))
ax3.bar(x,y, width=0.1, label='Action')
ax3.bar(x+0.1, y2, width=0.1, label='Adventure')
ax3.bar(x+0.2, y3, width=0.1, label='Fighting')
ax3.bar(x+0.3, y4, width=0.1, label='Misc')
ax3.bar(x+0.4, y5, width=0.1, label='Platform')
ax3.bar(x+0.5, y6, width=0.1, label='Puzzle')
ax3.bar(x+0.6, y7, width=0.1, label='Racing')
ax3.bar(x+0.7, y8, width=0.1, label='Role-Playing')
ax3.bar(x+0.8, y9, width=0.1, label='Shooter')
ax3.bar(x+0.9, y10, width=0.1, label='Simulation')
ax3.bar(x+1, y11, width=0.1, label='Sports')
ax3.bar(x+1.1, y12, width=0.1, label='Strategy')

# label dan legend
ax3.set_xticks(x+0.6) #gabisa langsung pake label, udah gabisa
ax3.set_xticklabels(list(labels))
ax3.set_xlabel('Regional Sales')
ax3.set_ylabel('Mean Sales')
ax3.set_title('Mean Sales Video Games by Genre', fontsize='xx-large')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# plt.style.use('seaborn')
# ax = datax.plot(kind='bar', figsize=(16,8), rot=0, title='Mean Sales Video Games by Genre')
# ax.set_xlabel('Region Sales')
# ax.set_ylabel('Mean Sales')
# plt.show()
````

Jawab: 
- Secara umum, penjualan tertinggi berada di region North America, lalu disusul oleh Eropa dan Jepang. Penjualan video games pada region lain tidak terlalu signifikan.

- Pada region NA, penjualan tertinggi dicapai oleh game bergenre platform. Top5 penjualan secara berurutan adalah genre platform, shooter, sports, racing, dan action. Penjualan game bergenre strategy dan shooter jauh melampaui genre lainnya.

- Pada region Eropa, penjualan tertinggi dicapai oleh game bergenre shooter. Top5 penjualan adalah genre shooter, platform, racing, sports, dan action. 

- Pada region Jepang, penjualan tertinggi dicapai oleh game bergenre role playing. Top5 penjualan adalah role playing, platform, fighting, puzzle, dan simulation. Penjualan genre role playing jauh melebihi genre lainnya

- Pada region lain, penjualan tiap genre tidak berbeda secara signifikan. Penjualan tertinggi dicapai oleh genre shooting dengan top5 penjualan adalah genre shooter, racing, platform, sports, dan action.

- Top 5 genre yang paling diminati pada pasar selain pasar Jepang adalah genre Action, Platform, Racing, Shooter, dan Sports. Dapat dikatakan bahwa secara global, 5 genre inilah yang paling diminati, sebab akumulasi penjualan pasar NA, Eropa, dan lainnya jauh melebihi penjualan pasar Jepang.

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

````python
#code here
x = datax.index
y = datax['Action']
y2 = datax['Adventure']
y3 = datax['Fighting']
y4 = datax['Misc']
y5 = datax['Platform']
y6 = datax['Puzzle']
y7 = datax['Racing']
y8 = datax['Role-Playing']
y9 = datax['Shooter']
y10 = datax['Simulation']
y11 = datax['Sports']
y12 = datax['Strategy']

# plot
fig4, ax4 = plt.subplots(figsize=(16,8))
ax4.bar(x, y, label='Action')
ax4.bar(x, y2, bottom=y, label='Adventure')
ax4.bar(x, y3, bottom=y+y2, label='Fighting')
ax4.bar(x, y4, bottom=y+y2+y3, label='Misc')
ax4.bar(x, y5, bottom=y+y2+y3+y4, label='Platform')
ax4.bar(x, y6, bottom=y+y2+y3+y4+y5, label='Puzzle')
ax4.bar(x, y7, bottom=y+y2+y3+y4+y5+y6, label='Racing')
ax4.bar(x, y8, bottom=y+y2+y3+y4+y5+y6+y7, label='Role-Playing')
ax4.bar(x, y9, bottom=y+y2+y3+y4+y5+y6+y7+y8, label='Shooter')
ax4.bar(x, y10, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9, label='Simulation')
ax4.bar(x, y11, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9+y10, label='Sports')
ax4.bar(x, y12, bottom=y+y2+y3+y4+y5+y6+y7+y8+y9+y10+y11, label='Strategy')
#legend
ax4.set_xlabel('Region Sales')
ax4.set_ylabel('Mean Sales')
ax4.set_title('Mean Sales Video Games by Genre')
fig4.legend(facecolor='inherit', framealpha=1)
plt.show()
````

Jawab: 
- Penjualan di region NA jauh melebihi penjualan di region manapun, disusul oleh Eropa dan Jepang
- Platform dan shooter adalah genre yang paling diminati di region NA, Eropa, dan lainnya. Sedangkan di region Jepang, genre yang paling diminati adalah role playing dan platform. Dapat dikatakan bahwa region Jepang memiliki selera yang cukup berbeda dibandingkan region lainnya.
