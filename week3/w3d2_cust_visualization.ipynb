<h2>Soal 1:</h2>

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

Expected Result:

![alt text](https://drive.google.com/uc?id=1XRcfWyiWLDmriw7s7U4fTn5Gs6s2qf5o)

```python
# code here
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(x,cos, color='r', label='Cosx')
ax.plot(x,cos3, linestyle='dashed', color='g', label='3 Cosx')
ax.plot(x,sin, linestyle='-.', color='b', label='Sin(x-90)')
ax.plot(x,sin2, marker='x', linestyle='None', color='y', label='2 Sinx')
# legend
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
ax.legend(loc='center')
plt.show()
```
